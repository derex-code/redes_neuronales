1. Tensors in PyTorch

```python
#PyTorch to build a simple neural network
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import torch

import helper
```

```python
#How we work with Pytorch tensors

x = torch.rand(3, 2)
x
```

```python
y = torch.ones(x.size())
y
```

```python
z = x + y
z
```

```python
z[0]
```

```python
z[:, 1:]
```

```python
# Return a new tensor z + 1
z.add(1)
```

```python
# z tensor is unchanged
z
```

```python
# Add 1 and update z tensor in-place
z.add_(1)
```

```python
# z has been updated
z
```
